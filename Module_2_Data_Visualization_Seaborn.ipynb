```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mtcars = pd.read_csv('data/mtcars.csv',
                     index_col='car_names')
mtcars.head()
```

```python
sns.set(style="darkgrid")
# sns.set_style("whitegrid")

plt.scatter(mtcars.mpg, mtcars.hp) # Only if no space in between
# plt.scatter(mtcars['mpg'], mtcars['hp'])
plt.xlabel('mpg')
plt.ylabel('hp')
plt.title('mtcars')
plt.show()
```

## Visualizing Statistical Relationships Between Quantiative Data

## Scatter Plot

```python
sns.scatterplot(x='mpg', y="hp", data=mtcars)
plt.show()
```

```python
fig = plt.figure(figsize=(6, 6), facecolor='w')
sns.scatterplot(x="mpg", y="hp", data=mtcars)
fig.savefig('mtcar', dpi=200, bbox_inches="tight")
```

```python
sns.relplot(x="mpg", y="hp", data=mtcars)
plt.show()
```

```python
sns.relplot(x="mpg", y="hp", data=mtcars, kind = 'scatter')
plt.show()
```

```python
sns.relplot(x="mpg", y="hp", data=mtcars, kind = 'line')
plt.show()
```

```python
sns.relplot(x="mpg", y="hp",data=mtcars, hue="am")
plt.show()
```

```python
mtcars['am'] = mtcars['am'].astype(str)
mtcars.info()
```

```python
mtcars['am'] = mtcars['am'].astype(int)
mtcars.info()
```

```python
sns.relplot(x="mpg", y="hp",data=mtcars, hue="cyl")
plt.show()
```

```python
# mtcars['cyl'] = mtcars['cyl'].astype(str)
# mtcars['cyl'] = mtcars['cyl'].astype(int)
mtcars['cyl'] = mtcars['cyl'].astype(float)
```

```python
sns.relplot(x="mpg", y="hp", size="cyl",data=mtcars);
```

```python
sns.relplot(x="mpg", y="hp", size="cyl", hue = 'cyl', data=mtcars);
```

```python
sns.relplot(x="mpg", y="hp", size="cyl",sizes=(20, 200), data=mtcars);
```

```python
sns.scatterplot(x="mpg", y="hp", hue="cyl", size="cyl",sizes=(20, 200),data=mtcars);
```

```python
sns.relplot(x="mpg", y="hp", col="cyl", size="cyl",sizes=(20, 200),data=mtcars);
```

```python
cmap = sns.cubehelix_palette(dark=.5, light=0.9, as_cmap=True)
sns.relplot(x="mpg", y="hp", hue="cyl", size="cyl",sizes=(20, 200),palette=cmap,data=mtcars)
```

```python
sns.relplot(x="mpg", y="hp", hue="cyl", size="cyl",sizes=(20, 200),
            style='am', palette='Set2',data=mtcars);
```

```python
cmap = sns.cubehelix_palette(dark=.5, light=0.9, as_cmap=True)
sns.scatterplot(x="mpg", y="hp", hue="cyl", size="cyl",sizes=(20, 200),style='am',data=mtcars);
```
