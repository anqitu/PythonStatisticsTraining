# Module 5 Bayesian Inference

## Demo Bayesian Inference


```
import numpy as np
import pandas as pd

x = np.linspace(-5, 5, 20)
y = -5 + 3*x + np.random.normal(size=x.shape)
data = pd.DataFrame({'x': x, 'y': y})
```

```
data
```

```
import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.figure()
ax = fig.add_subplot()
ax.grid(False)
ax.spines['bottom'].set_position('zero')
ax.spines['left'].set_position('zero')

plt.plot(x,y,'or')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```
import pymc3 as pm
```

```
# creates a PyMC model object
# Establishes a context manager. 
# All PyMC objects created within the context manager are added to the model object.
with pm.Model():
    
    # prior distribution 
    intercept = pm.Uniform('intercept',-10,0)
    slope = pm.Uniform('slope',1,3)
    
    # likelihood (Observation)
    obs = pm.Normal('y', 
                   mu = intercept+slope*x, # prior distribution 
                   sd = 1,
                   observed = y) # obervation
    
    #Sampling distribition of outcomes in the dataset.
    #our prior distribution (intercept & slope) will be updated with data
    trace = pm.sample(1000)
```

```
# !pip install git+https://github.com/pymc-devs/pymc3
```

```
# !pip install git+git://github.com/arviz-devs/arviz.git
```

```
pm.traceplot(trace)
```

```
# Now that we have a full distribution for the probability of various values of θ, 
# we can take the mean of the distribution as our most plausible value for θ

a = trace['intercept'].mean()
b = trace['slope'].mean()

yhat = a + b*x

import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(x,y,'or')
plt.plot(x,yhat,'b')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
